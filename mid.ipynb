```python
from scipy.stats import binom

# Parametry
n = 9600
p = 0.03

# Chcemy policzyć P(X >= 350).
# CDF(k) zwraca prawdopodobieństwo P(X <= k).
# Dlatego używamy k=349, żeby "odciąć" lewą stronę wykresu (0 do 349).
prob = 1 - binom.cdf(349, n, p)

print(f"Prawdopodobieństwo (1 - cdf): {prob}")
print(f"W notacji dziesiętnej: {prob:.8f}")
```

```python
import kagglehub
import pandas as pd
import numpy as np
from scipy.stats import norm
from math import ceil

# Download the latest version
path = kagglehub.dataset_download("dellakoovakkattu/bellabeatdailyactivity")

import os
for file in os.listdir(path):
    print(file)
df = pd.read_csv(os.path.join(path, 'dailyActivity_merged.csv'))
df.columns

df_cleaned = df['VeryActiveMinutes'].dropna()
df_cleaned

x = df["VeryActiveMinutes"].dropna()

n = len(x)
x_bar = x.mean()
s = x.std(ddof=1)
print(x_bar)
se = s / np.sqrt(n)
print(se)

# ----------------------------------------------------
# 99% confidence interval for the mean
# ----------------------------------------------------
z = norm.ppf(0.995)  # two-sided 99%
margin = z * se

ci_lower = x_bar - margin
ci_upper = x_bar + margin

print(ci_lower, ci_upper)
print(margin)

relative_error_percent = (margin / x_bar) * 100

print(relative_error_percent)

desired_precision = 0.03 * x_bar
n_required = ceil((z * s / desired_precision) ** 2)
print(n_required)

```

```python
from scipy.stats import expon, poisson

more_than_25 = expon.sf(25, scale=12)
print(more_than_25)


mean_time = 12          # średnio co 12 minut
duration = 2 * 60       # 2 godziny = 120 minut

# SPOSÓB 1: Exponential Distribution
# Pytanie: Jakie jest szansa, że czas do zdarzenia > 120?
# Używamy Survival Function (sf)
prob_expon = expon.sf(duration, scale=mean_time)

# SPOSÓB 2: Poisson Distribution (dla sprawdzenia)
# Oczekiwana liczba zdarzeń (lambda) w tym czasie: 120 / 12 = 10 zdarzeń
expected_events = duration / mean_time
# Pytanie: Prawdopodobieństwo, że liczba zdarzeń (k) wyniesie 0
prob_poisson = poisson.pmf(0, mu=expected_events)

print(f"Prawdopodobieństwo (Expon):   {prob_expon:.10f}")
print(f"Prawdopodobieństwo (Poisson): {prob_poisson:.10f}")
print(f"W notacji naukowej: {prob_expon:.2e}")
```
