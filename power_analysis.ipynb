```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean1 = 100
std = 16
n = 60
mean2 = 108


# generate sample draws (optional) and plot theoretical PDFs
rng = np.random.default_rng(0)
s1 = rng.normal(mean1, std, n)
s2 = rng.normal(mean2, std, n)

x = np.linspace(min(mean1, mean2) - 4*std, max(mean1, mean2) + 4*std, 500)
pdf = lambda mu: (1/(std * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x - mu)/std)**2)

plt.figure(figsize=(8,5))
low_cut = 85
high_cut = 115
plt.plot(x, pdf(mean1), label=f"N({mean1}, {std}^2)", color='C0')
plt.plot(x, pdf(mean2), label=f"N({mean2}, {std}^2)", color='C1')
plt.xlim(low_cut, high_cut)
z95 = 1.65
# scalar critical value
xcrit_scalar = mean1 + z95 * std / np.sqrt(n)
plt.axvline(xcrit_scalar, color='k', linestyle='--', label=f'critical = {xcrit_scalar:.2f}')
plt.xlabel('x'); plt.ylabel('Density')
plt.title('Two normal distributions (same std, different means)')
plt.legend()
plt.show()
plt.xlabel('x'); plt.ylabel('Density')
plt.title('Two normal distributions (same std, different means)')
plt.legend()
plt.show()
xcrit_scalar


z = (xcrit_scalar - mean2) / (std / np.sqrt(n))
prob = 1 - norm.cdf(z)
prob
```
